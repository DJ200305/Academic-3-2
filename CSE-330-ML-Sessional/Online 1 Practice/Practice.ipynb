Exercise 1: Handling Missing Data and Filtering

```python
import pandas as pd
import numpy as np

# Sample Data: Sales records with some missing values
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': [2500, np.nan, 4500, 3100, np.nan, 5200],
    'Quantity': [10, 5, np.nan, 8, 12, 15]
}
df = pd.DataFrame(data)

# TODO: Check which cells contain missing values (NaN)
missing_mask = df.isnull().sum()
print(missing_mask)
# TODO: Fill the missing values in 'Sales' and 'Quantity' using the mean of their respective columns

df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
print(df.isnull().sum())
# TODO: Filter the dataframe to show only rows where 'Sales' is greater than 3000
df_filtered = df.query('Sales > 3000')

print(df_filtered)
```

Exercise 2: Feature Scaling (Min-Max Normalization)

```python
import pandas as pd
import numpy as np

# Sample Data: House prices and square footage
data = {
    'Price': [300000, 450000, 200000, 800000, 550000],
    'SqFt': [1500, 2100, 1200, 3500, 2500]
}
df = pd.DataFrame(data)

# TODO: Convert the 'Price' column into a NumPy array
price_array = np.array(df['Price'])
print(price_array)
# TODO: Calculate the Min-Max Normalization for the 'Price' array using NumPy functions
# (Scale the values between 0 and 1)

price_min = np.min(price_array)
price_max = np.max(price_array)
normalized_price = np.array((price_array-price_min)/(price_max-price_min))

# TODO: Add the normalized prices back to the dataframe as a new column 'Price_Scaled'
df['Price_Scaled'] = pd.DataFrame(normalized_price)

print(df)
```

Exercise 3: Categorical Encoding and Data Aggregation

```python
import pandas as pd
import numpy as np

# Sample Data: Employee departments and performance scores
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Dept': ['HR', 'Tech', 'Tech', 'HR', 'Design'],
    'Score': [85, 92, 78, 88, 95]
}
df = pd.DataFrame(data)
print(df)
# TODO: Use Pandas to perform One-Hot Encoding on the 'Dept' column
# Hint: Use pd.get_dummies()
df_encoded = pd.get_dummies(df['Dept'])
print("After encoding:")
print(df)
# TODO: Group the original dataframe by 'Dept' and find the average 'Score' for each department
dept_avg = df.groupby(['Dept']).mean(numeric_only=True)

print("Encoded Dataframe:")
print(df_encoded)
print("\nDepartment Averages:")
print(dept_avg)
```

Exercise 4: Detecting Outliers with NumPy (Z-Score)

```python
import pandas as pd
import numpy as np

# Sample Data: Sensor readings with some extreme noise
data = {
    'Sensor_ID': range(1, 11),
    'Reading': [25.1, 24.8, 25.3, 100.2, 25.0, 24.9, -50.5, 25.2, 25.1, 25.0]
}
df = pd.DataFrame(data)

# TODO: Calculate the mean and standard deviation of the 'Reading' column using NumPy
reading_mean = np.mean(np.array(df['Reading']))
reading_std = np.std(np.array(df['Reading']))

# TODO: Calculate the Z-score for every value in 'Reading'
# Formula: (value - mean) / std
z_scores = ((df['Reading']-reading_mean)/reading_std)

# TODO: Create a boolean mask where Absolute Z-score is less than 3
# Hint: Use np.abs()
mask = np.abs(z_scores) < 3
print(mask)
# TODO: Use the mask to create a cleaned dataframe without outliers
df_cleaned = df[mask].copy()
# df_cleaned.drop('Reading_Outlier')
print("Cleaned Data:")
print(df_cleaned)
```

Exercise 5: Advanced Pivot and Data Binning

```python
import pandas as pd
import numpy as np

# Sample Data: Customer ages and spending scores
np.random.seed(42)
data = {
    'Age': [22, 45, 33, 54, 61, 18, 25, 41, 32, 55],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    'Spend_Score': [10, 88, 40, 95, 20, 15, 30, 70, 45, 60]
}
df = pd.DataFrame(data)

# TODO: Create Age Groups (Bins)
# Create 3 bins: 0-30 (Young), 30-50 (Middle), 50-100 (Senior)
# Use pd.cut() with labels=['Young', 'Middle', 'Senior']
df['Age_Group'] = pd.cut(np.array(df['Age']),3,labels=['Young', 'Middle', 'Senior'])
print(df)
# TODO: Create a Pivot Table
# Index should be 'Age_Group', Columns should be 'Gender', and Values should be 'Spend_Score' (mean)
pivot_results = df.pivot_table(index='Age_Group',columns='Gender',values='Spend_Score')

print(pivot_results)
```

Exercise 6: Vectorized String Operations and Multi-Condition Logic

```python
import pandas as pd
import numpy as np

# Sample Data: Product codes and stock levels
data = {
    'Product_Code': ['ELEC_101', 'FURN_202', 'ELEC_105', 'OFFC_303', 'FURN_205'],
    'Base_Price': [500, 200, 1200, 50, 300],
    'Quantity': [5, 0, 2, 100, 1]
}
df = pd.DataFrame(data)

# TODO: Extract the Category (the text before the underscore) from 'Product_Code'
# Hint: Use .str.split().str[0]
df['Category'] = df['Product_Code'].str.split().str[0]

# TODO: Apply a complex discount logic using np.select()
# 1. If Category is 'ELEC' and Base_Price > 1000, Discount = 0.2
# 2. If Category is 'FURN' or Quantity == 0, Discount = 0.1
# 3. Otherwise, Discount = 0.05
conditions = [
    # Your code here (Condition 1),
    (df['Category'] == 'ELEC') & (df['Base_Price'] > 1000),
    (df['Category'] == 'FURN') | (df['Quantity'] == 0)      
    # Your code here (Condition 2)
]
choices = [0.2, 0.1]
df['Discount'] = np.select(conditions, choices, default=0.05)

# TODO: Calculate Final_Price = Base_Price * (1 - Discount)
df['Final_Price'] = df['Base_Price'] * (1-df['Discount'])

print(df)
```
